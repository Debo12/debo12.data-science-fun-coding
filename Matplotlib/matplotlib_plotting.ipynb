```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Average daily temperature in Jan 2018')
ax.set_xlabel('Day')
ax.set_ylabel('Average temperature (in degree)')
ax.set_xlim([0, 30])
ax.set_ylim(25, 35)
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
ax.plot(days, temp, label='Avg temp', color='green', linestyle='solid', linewidth=3, marker='o')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Average daily temperature in Jan 2018')
ax.set_xlabel('Day')
ax.set_ylabel('Average temperature (in degree)')
ax.set_xlim([0, 30])
ax.set_ylim(25, 35)
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
location1_temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
location2_temp = [26.4, 26.8, 26.1, 26.4, 27.5, 27.3, 26.9, 26.8, 27.0]
ax.plot(days, location1_temp, label='Avg temp 1', color='green', linestyle='solid', linewidth=3, marker='o')
ax.plot(days, location2_temp, label='Avg temp 2', color='red', linestyle='solid', linewidth=3, marker='o')
plt.legend()
plt.show()
```

# Scatter Plot

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Average daily temperature in Jan 2018')
ax.set_xlabel('Day')
ax.set_ylabel('Average temperature (in degree)')
ax.set_xlim([0, 30])
ax.set_ylim(25, 35)
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
location1_temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
location2_temp = [26.4, 26.8, 26.1, 26.4, 27.5, 27.3, 26.9, 26.8, 27.0]
ax.scatter(days, location1_temp, label='Location1', s=[60], c='green', marker='o', edgecolors='black')
ax.scatter(days, location2_temp, label='Location2', s=[60], c='orange', marker='^', edgecolors='black')
plt.legend()
plt.show()
```
